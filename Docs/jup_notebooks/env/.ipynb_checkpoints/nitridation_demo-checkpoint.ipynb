We then import all the useful modules, including mutationpp and sofia with the sys.path.append method.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.optimize import minimize
from sklearn import gaussian_process
from sklearn.gaussian_process.kernels import Matern, RBF
import seaborn as sns
import pickle
import json
import sys
import os
sys.path.append('../../../installation/build/lib') # Your path to sofia
sys.path.append('/Library/Frameworks/Python.framework/Versions/3.6/lib/python3.6/site-packages/mutationpp-0.0.1-py3.6-macosx-10.6-intel.egg') # Your path to the mutationpp .egg file
import sofia.sampler as mcmc
import sofia.distributions as dist
import sofia.data_assembly as assemble
import mutationpp as mpp
```

As the json file containing the data to find the different models in SoFIA/Applications/nitridation/models only contains a relative path (for simplicity), this class of functions allows us to change the working directory so that it find the files even when pointed to with a relative path.

```python
class cd:
    """Context manager for changing the current working directory""" # Needed for the handling of models files
    def __init__(self, newPath):
        self.newPath = os.path.expanduser(newPath)

    def __enter__(self):
        self.savedPath = os.getcwd()
        os.chdir(self.newPath)

    def __exit__(self, etype, value, traceback):
        os.chdir(self.savedPath)
```

We first set-up the mutation mixture and options.

```python
## Setting up Mutation++ options and mixture
opts = mpp.MixtureOptions("nitrogen-ions-carbon_9_olynick99")
opts.setThermodynamicDatabase("RRHO")
opts.setStateModel("ChemNonEq1T")

mix = mpp.Mixture(opts)
```

Subsequently, we load the models and data. The sys.argv list of strings emulates what you would have to write on the command line if you were to run this code in a script.

```python
## Loading models and data

sys.argv = ['script name','G4','1T','SEB']

sofia_dir = '/Users/anabel/Documents/PhD/Code/' # Directory wherever you have the SoFIA library

with cd(sofia_dir + "SoFIA/Applications/nitridation"): # We change the working directory to this one just to instatiate the object for the assembly of the priors and likelihood
    assembly = assemble.assembly('./cases.json','./models/models.json',mix)
measurements = ["Ps","Pd","Tw","recession"]
assembly.lik_hyperparams(measurements)
assembly.assembly_prior()
```

We use an optimization algorithm within scipy to find the log-likelihood maximum and use it to start our MCMC sampling.

```python
## Looking for the MAP point to start sampling ##
Xi = [0.5]*len(assembly.hyp)
res = scipy.optimize.minimize(assembly.m_log_likelihood,Xi,method='Nelder-Mead',tol=1e-6)
print("MAP found at: "+str(res.x))
```

Sampling of the posterior via MCMC Metropolis-Hastings algorithm. The number of burned-in samples and the total number of samples that compose the final chain can be chosen.

```python
# MCMC sampling
nburn = 10000
sampler = mcmc.metropolis(np.identity(len(assembly.hyp))*0.01,assembly.log_likelihood,nburn)

sampler.seed(res.x)
sampler.Burn()

nchain = 100000
XMCMC = np.zeros((nchain,len(assembly.hyp)))

for i in range(nchain):
    XMCMC[i] = sampler.DoStep(1)
    print("MCMC step: "+str(i))
```

We save the resulting chain in an external file for posterior use.

```python
# Saving the chain to external file
filename = './chain_'+sys.argv[1]+'_'+sys.argv[2]+'_'+sys.argv[3]+'.dat'

with open(filename,'w') as ch:
    for i in range(len(XMCMC)):
        for j in range(len(assembly.hyp)-1):
            ch.write(str(XMCMC[i,j])+' ')
        ch.write(str(XMCMC[i,len(assembly.hyp)-1])+'\n')
```

We plot the marginal distributions of a handful of parameters such as $\gamma_{\mathrm{N}}^{\mathrm{CN}}, P_{\mathrm{s}}, P_{\mathrm{d}}$ and $T_{\delta}$, the temperature at the edge of the boundary layer, considered unknown in the analysis.

```python
## Marginals plotting
Xi_denorm = np.zeros((nchain,len(assembly.hyp)))

for i in range(nchain):
    Xi_denorm[i] =  assembly.denormalization(XMCMC[i,:])

plt.figure()
sns.kdeplot(Xi_denorm[:,assembly.models[assembly.therm_model][assembly.model]["indexes"]["Gnit"]],label='$\gamma_{\mathrm{N}}^{\mathrm{CN}}$',shade=True)
sns.kdeplot(Xi_denorm[:,assembly.models[assembly.therm_model][assembly.model]["indexes"]["Grec"]],label='$\gamma_{\mathrm{N}}^{\mathrm{N}_{2}}$',shade=True)
plt.xlim(-4.,0.)
plt.legend()

plt.figure()
sns.kdeplot(Xi_denorm[:,assembly.models[assembly.therm_model][assembly.model]["indexes"]["Ps"]],label='Ps',shade=True)
plt.legend()

plt.figure()
sns.kdeplot(Xi_denorm[:,assembly.models[assembly.therm_model][assembly.model]["indexes"]["Pd"]],label='Pd',shade=True)
plt.legend()

plt.figure()
sns.kdeplot(Xi_denorm[:,assembly.models[assembly.therm_model][assembly.model]["indexes"]["Te"]],label='Te',shade=True)
plt.legend()
```

Lastly, we compute some diagnostics, such as the trace and autocorrelation function.

```python
## Chain diagnostics
dict_var={0:'$\gamma_{\mathrm{N}}^{\mathrm{CN}}$', 1: '$\gamma_{\mathrm{N}}^{\mathrm{N}_{2}}$'} # The way they are positioned in MCMC chain

var = [0, 1] # Position in XMCMC chain

sampler_diag = mcmc.diagnostics(XMCMC,dict_var)
sampler_diag.chain_visual(2,var,10000)
sampler_diag.autocorr(100,2,var)
```
